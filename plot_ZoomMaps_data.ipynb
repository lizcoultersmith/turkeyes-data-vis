```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
import visfunctions as sp
import utilfunctions as sf
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
def visualize_attention_zoom(imdir, impdir, plotN=5, plottype=2):
    # plottype:
    # 1 for heatmap only
    # 2 for heatmap overlay
    # 3 for regular spotlight
    # 4 for spotlight with level sets
    # SIGMA is used for converting locations into maps (distributions)
    # plotN is how many images to plot
    imnames = [file for file in os.listdir(imdir) if not file.startswith('.')]
    
    if plotN:
        imnames = imnames[:plotN] 

    for filename in imnames:
        impim, im = sf.get_processed_heatmap_for_zoom(filename, imdir, impdir)
        sp.plot_attention(filename, im, impim, plottype)
```
